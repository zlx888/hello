# [for 循环语句（for loops）](https://docs.python.org/3/tutorial/controlflow.html#for-statements)

Python for循环可以遍历任何序列的项目，如一个列表或者一个字符串。

for循环的语法格式如下：

```python
for iterating_var in sequence:
   statements(s)
```

````python
my_list = [1, 2, 3, 4, 'Python', 'is', 'neat']
for item in my_list:
    print(item)
````

- `break` 可以终止整个循环语句的运行

````python
for item in my_list:
    if item == 'Python':
        break
    print(item)
````

- `continue` 语句跳出本次循环，而 break 跳出整个循环。

continue 语句用来告诉Python跳过当前循环的剩余语句，然后继续进行下一轮循环。

continue语句用在 while 和 for 循环中。

````python
for item in my_list:
    if item == 1:
        continue
    print(item)
````

- `enumerate()` 可以返回 element 的 index

````python
for idx, val in enumerate(my_list):
    print('idx: {}, value: {}'.format(idx, val))
````

一个常用的函数 `range()` 可以返回一个序列，所以它常常被用在 for-loop 中：

````python
for number in range(5):
    print(number)
````

````python
for number in range(2, 5):
    print(number)
````

````python
for number in range(0, 10, 2):  # last one is step
    print(number)
````

因此，我们也可以通过 `range()` 函数，通过遍历一个 list 的 index：

````python
for idx in range(len(my_list)):
    print('idx: {}, value: {}'.format(idx, my_list[idx]))
````

## 循环遍历字典中的元素

````python
my_dict = {'hacker': True, 'age': 72, 'name': 'John Doe'}
for val in my_dict:
    print(val)
````

````python
for key, val in my_dict.items():
    print('{}={}'.format(key, val))
````

## *While 循环

下面是一个 while 循环的例子，其语法为：

```python
while 判断条件：
    执行语句……
```

当判断条件假false时，循环结束。

Gif 演示 Python while 语句执行过程:

![](imgs/while-loop-demo.gif)

````python
var = 10
while var > 0:              
   var = var -1
   if var == 5:
      continue
   print('当前变量值 :', var)
````
