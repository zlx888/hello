# Notebook 用户
- 学生及自学者
    - Notebook 可以被视为一本电子书
    - 书中的内容可以随时交互
- Notebook 作者
    - 为读者创建 Notebook

# Notebook 的基本单元是 cell
- Code cells
    - 运行和计算代码块 (Python 及其他程序，包括 SAS)
- Markdown cells
    - 用于编写文档和注解
        - 引导读者理解
