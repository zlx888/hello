# [Modules and packages]

> 一个 .py 后缀的文件就是一个模块

> 包是一个包含有 `__init__.py` 文件的目录，其中可以包含模块和其他包  


## 包和模块组织代码的好处
* 便于维护
* 可重用
* 命名空间隔离
* 通过文件夹结构就可以清楚浏览项目
* 方便查找函数和类

## 如何使用

以以下目录结构为例:

      
```
food_store/
    __init__.py
    
    product/
        __init__.py
        
        fruit/
            __init__.py
            apple.py
            banana.py
            
        drink/
            __init__.py
            juice.py
            milk.py
            beer.py

    cashier/
        __ini__.py
        receipt.py
        calculator.py
```


以上 banana.py 文件内容：

```python

def get_available_brands():
    return ['chiquita']


class Banana:
    def __init__(self, brand='chiquita'):
        if brand not in get_available_brands():
            raise ValueError('Unkown brand: {}'.format(brand))
        self._brand = brand
     
```

### Importing

如果我们要在receipt.py文件中访问banana.py模块中的 `Banana` 类。我们需要在receipt.py开始的时候导入该类:

```python
from food_store.product.fruit.banana import Banana

# 如下方式使用
my_banana = Banana()
```



If we need to access multiple classes or functions from banana.py file:

```python
from food_store.product.fruit import banana

# then it's used like this
brands = banana.get_available_brands()
my_banana = banana.Banana()
```
