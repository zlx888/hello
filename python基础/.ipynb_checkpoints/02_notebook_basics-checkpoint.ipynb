# Notebook 用户界面
- 菜单栏
- 工具栏
- Notebook 区（内含 cells 单元格）

# 交互中 cell 编辑器
- Command Mode
    - 蓝色 cell 边框
    - 通过 'ESC' key 进入 command mode
- Edit Mode
    - 绿色 cell 边框
    - cell 相当于一个代码编辑器
    - 通过回车键进入 edit mode
- 运行 cell
    - 点击工具栏运行按钮，或使用 'SHIFT ENTER' 快捷键

# 快捷键\*
- command palette tool icon
    - has search capability

### 常用导航功能 
- 保存 notebook: 'CTRL' + s
- Create cell above current cell: 'ESC' + a
- Create cell below current cell: 'ESC' + b
- 删除当前 cell: 'ESC' + d + d
- Cell 编辑
    - copy: 'CTRL' + c
    - cut: 'CTRL' + x
    - paste: 'CTRL' + v
    - undo: 'CTRL' + z
