# [数字（Numbers）](https://docs.python.org/3/library/stdtypes.html#numeric-types-int-float-complex)

## `int`

整型

```python
my_int = 6
print('value: {}, type: {}'.format(my_int, type(my_int)))
```

## `float`

浮点型

```python
my_float = float(my_int)
print('value: {}, type: {}'.format(my_float, type(my_float)))
```

注意：两个`int` 相除会得到一个 `float`:

```python
print(1 / 1)
print(6 / 5)
```

## Floor division `//`, modulus `%`, power `**`

7 除以 5 的商数：

```python
7 // 5
```

7 除以 5 的余数：

```python
7 % 5
```

2 的 3 次方：

```python
2 ** 3
```

## 数学运算
在 Python 中，数学计算的运算顺序与代数相同。可以使用括号来改变部分运算步骤的优先级。

```python
print(1 + 2**2 * 3 / 6) # 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```
